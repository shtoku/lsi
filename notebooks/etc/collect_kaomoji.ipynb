# 顔文字の収集

```python
from bs4 import BeautifulSoup
import requests

import time
import random
```

## 「顔文字BOX」から収集
https://kmoji.com

```python
def collect_kmoji(num):
  url = "https://kmoji.com/category/" + str(num)

  res = requests.get(url)
  soup = BeautifulSoup(res.content, 'html.parser')

  el = soup.find('div', attrs={ 'class': 'archive' }).find_all('a')

  return el
```

```python
with open('kmoji.txt', mode='w') as f:
  for i in range(190):
    try:
      el = collect_kmoji(i+1)
      for a in el:
        f.write(a.text + '\n')
      time.sleep(1 + random.random())
    except:
      pass
```

## 「顔文字コピペ」から収集
https://kaomoji-copy.com

```python
base_url = "https://kaomoji-copy.com"

res = requests.get(base_url)
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
def collect_kaomoji_copy(url):
  res = requests.get(url)
  soup = BeautifulSoup(res.content, 'html.parser')

  el = soup.find('table',  attrs={ 'class': 'kaomoji' }).find_all('input')

  return el
```

```python
tags = soup.find('div', attrs={ 'class': 'kaomoji_categorywrap' }).find_all('a')

with open('kaomoji_copy.txt', mode='w') as f:
  for tag in tags:
    el = collect_kaomoji_copy(base_url + tag.get('href'))
    for i in el:
      val = i.get('value')
      f.write(val)
    time.sleep(1 + random.random())
```
